```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python

sentences = [
    "The batsman hit a magnificent six over the long-on boundary.",
    "Spinners often struggle to find grip on a dew-laden outfield.",
    "The test match ended in a thrilling draw after five days of play.",
    "Fielding at gully requires incredibly fast reflexes and soft hands.",
    # Topic: Cooking
    "Always sauté the onions until they turn translucent and golden.",
    "A pinch of saffron can elevate the aroma of any rice dish.",
    "Slow-cooking the meat ensures a tender and juicy texture.",
    # Topic: Cybersecurity
    "Multi-factor authentication adds a critical layer of account security.",
    "Phishing attacks often use deceptive emails to steal user credentials.",
    "End-to-end encryption protects data from being intercepted during transit."
]
model=SentenceTransformer('all-MiniLM-L6-v2')
embeddings=model.encode(sentences)
print(f"Embedding shape: {embeddings.shape}")
```

```python
similarity_matrix = cosine_similarity(embeddings)
query = ["The bowler took three wickets in one over"]
query_embedding = model.encode(query)
query_sims = cosine_similarity(query_embedding, embeddings)[0]

# Get indices of top 2 scores
top_indices = np.argsort(query_sims)[-2:][::-1]

print(f"Query: {query[0]}\n")
print("Top 2 most similar sentences:")
for idx in top_indices:
    print(f"- Score: {query_sims[idx]:.4f} | Sentence: {sentences[idx]}")
```
